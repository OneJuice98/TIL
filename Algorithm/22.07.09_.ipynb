### 덱 #10866
* deque : 양방향 큐, deque!
 1. 앞, 뒤 양쪽 방향에서 element를 추가 및 제거가능!
 2. append, pop이 압도적으로 빠름!
 3. 일반적인 list 는 O(n) 이지만 이친구는 O(1)의 복잡도래!
 
* deque method
 1. deque.append(item) : item을 deque의 오른쪽 끝에 삽입
 2. deque.appendleft(item) : item을 deque의 왼쪽 끝에 삽입
 3. deque.pop() : deque의 오른쪽 끝 element를 가져오는 동시에 deque에서 삭제
 4. deque.popleft() : deque의 왼쪽 끝 element를 가져오는 동시에 deque에서 삭제
 5. deque.extend(array) : 주어진 배열(array)을 순환하면서 deque의 오른쪽에 추가
 6. deque.extendleft(array) : 주어진 배열(array)을 순환하면서 deque의 왼쪽에 추가
 7. deque.remove(item) : item을 deque에서 찾아 삭제
 8. deque.rotate(num) : deque을 num만큼 회전한다.(양수 오른쪽, 음수 왼쪽)

```python
from collections import deque

N = int(input())

cmd_list = []
for i in range(N):
    cmd = input().split()
    if len(cmd) == 1:
        cmd.append('')
    cmd_list.append(cmd)

deque = deque()
for cmd in cmd_list:
    if cmd[0] == 'push_back':
        deque.append(int(cmd[1]))
    elif cmd[0] == 'push_front':
        deque.appendleft(int(cmd[1]))
    elif cmd[0] == 'pop_front':
        if len(deque) == 0:
            print(-1)
        else:
            print(deque.popleft())
    elif cmd[0] == 'pop_back':
        if len(deque) == 0:
            print(-1)
        else:
            print(deque.pop())
    elif cmd[0] == 'size':
        print(len(deque))
    elif cmd[0] == 'empty':
        if len(deque) == 0:
            print(1)
        else:
            print(0)
    elif cmd[0] == 'front':
        if len(deque) == 0:
            print(-1)
        else:
            print(deque[0])
    elif cmd[0] == 'back':
        if (len(deque) == 0):
            print(-1)
        else:
            print(deque[len(deque)-1])
```

### 큐 #10845
 * 선입선출(FIFO, First In First Out)
 * list append, pop, insert(idx, val)
 * 마찬가지로 pop(0)은 빼는 동시에 출력까지!

```python
N = int(input())

cmd_list = []
for i in range(N):
    cmd = input().split()
    if len(cmd) == 1:
        cmd.append('')
    cmd_list.append(cmd)

queue = []
for cmd in cmd_list:
    if cmd[0] == 'push':
        queue.append(cmd[1])
    elif cmd[0] == 'pop':
        if len(queue) == 0:
            print(-1)
        else:
            print(queue.pop(0))
    elif cmd[0] == 'size':
        print(len(queue))
    elif cmd[0] == 'empty':
        if len(queue) == 0:
            print(1)
        else:
            print(0)
    elif cmd[0] == 'front':
        if len(queue) == 0:
            print(-1)
        else:
            print(queue[0])
    elif cmd[0] == 'back':
        if len(queue) == 0:
            print(-1)
        else:
            print(queue[len(queue)-1])
```

### 수 정렬하기2 #2751
 * Quick Sort
  1. 입력된 자료 리스트에서 하나의 원소를 고른다. 이 원소를 피벗이라고 부른다
  2. 피벗을 기준으로 리스트를 둘로 분할한다.
  3. 피벗을 기준으로 피벗보다 작은 원소들은 모두 피벗의 왼쪽으로 옮긴다.
  4. 피벗을 기준으로 피벗보다 큰 원소들은 모두 피벗의 오른쪽으로 옮긴다.
  <br>
  → Quick Sort 는 python list sort method 와 동일, 되도록 사용하지 않도록!
 * Merge Sort
  1. 리스트의 요소가 1개가 될 때까지 나눈다.
  2. 분리한 왼쪽리스트, 오른쪽 리스트의 각각 첫번째 요소를 비교해 더 작은 값을 결과 리스트에 저장한다.
  3. 저장한 값은 리스트에서 지운다.
  4. 두 리스트 모두 요소가 하나도 안남을 때까지 반복한다.
  <br>
  → 시간 초과,,, 날 너무 강하게 만들려고 하네,,, 
 * Heap Sort
  1. Heap Tree 구성하여 정렬
  2. merge/quick 만큼 빠른 정렬 알고리즘!
  3. 내림차순 정렬시 최대힙 / 오름차순 최소힙
 * Heap : 최대값 및 최소값을 찾아내는 연산을 빠르게 하기 위해 고안된 완전이진트리의 자료구조
 * 최대힙 : 부모노드의 키값이 자식노드의 키값보다 항상 큰 경우
 * 최소힙 : 부모노드의 키값이 자식노드의 키값보다 항상 작은 경우
  1. 주어진 원소로 최대 힙 구성
  2. 최대 힙의 부모노드(배열 첫번째 요소) 와 말단노드(배열 마지막 요소) 교환
  3. 새 부모노드에 대해 최대힙 구성
  4. 원소의 개수만큼 2번, 3번 반복

```python
# Merge-Sort
def divid_arr(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    leftArr = divid_arr(arr[:mid])
    rightArr = divid_arr(arr[mid:])
    return merge_sort(leftArr, rightArr)
    
def merge_sort(left, right):
    arr = []
    cnt_l = cnt_r = 0
    while cnt_l <len(left) and cnt_r <len(right):
        if (left[cnt_l] < right[cnt_r]):
            arr.append(left[cnt_l]) # left가 더 클때 left 넣으면 내림차순, 오름차순은 반대
            cnt_l += 1
        else:
            arr.append(right[cnt_r])
            cnt_r += 1
    arr += left[cnt_l:]
    arr += right[cnt_r:]
    return arr
```

```python
N = int(input())

seq = []
for i in range(N):
    seq.append(int(input()))

def heapify(unsorted, index, heap_size):
    largest = index
    left_index = 2 * index + 1
    right_index = 2 * index + 2
    if left_index < heap_size and unsorted[left_index] > unsorted[largest]:
        largest = left_index
    if right_index < heap_size and unsorted[right_index] > unsorted[largest]:
        largest = right_index
    if largest != index:
        unsorted[largest], unsorted[index] = unsorted[index], unsorted[largest]
        heapify(unsorted, largest, heap_size)

def heap_sort(unsorted):
    n = len(unsorted)
    # BUILD-MAX-HEAP (A) : 위의 1단계
    # 인덱스 : (n을 2로 나눈 몫-1)~0
    # 최초 힙 구성시 배열의 중간부터 시작하면 
    # 이진트리 성질에 의해 모든 요소값을 
    # 서로 한번씩 비교할 수 있게 됨 : O(n)
    for i in range(n // 2 - 1, -1, -1):
        heapify(unsorted, i, n)


    # Recurrent (B) : 2~4단계
    # 한번 힙이 구성되면 개별 노드는
    # 최악의 경우에도 트리의 높이(logn)
    # 만큼의 자리 이동을 하게 됨
    # 이런 노드들이 n개 있으므로 : O(nlogn)
    for i in range(n - 1, 0, -1):
        unsorted[0], unsorted[i] = unsorted[i], unsorted[0]
        heapify(unsorted, 0, i)
    return unsorted
        
ans = heap_sort(seq)
for i in range(ans):
    print(i)
```
